```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV dataset into a pandas DataFrame
data = pd.read_csv("train.csv")
# Display the first 5 rows of the dataset
data.head()
```

```python
# Show the column names of the dataset
data.columns
```

**Survived**

```python
# Look at the distribution of the 'Survived' column (target variable)
survivors = data['Survived'].value_counts()
survivors
```

```python
# Plot a countplot to visualize the number of survivors (1) and non-survivors (0)
sns.countplot(x='Survived', data=data)
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

SEX

```python
# Count the number of male and female passengers in the dataset
sex_count_values =data['Sex'].value_counts()
sex_count_values
```

```python
# Create a bar plot to show the distribution of passengers by sex (Male/Female)
plt.bar(data['Sex'].value_counts().index, data['Sex'].value_counts().values)
plt.xlabel('Sex')
plt.ylabel('Number of passengers')
plt.title('Sex distribution')
plt.show()
```

```python
# Count the number of dead male passengers (Survived == 0 and Sex == 'male')
male_dead = ((data['Sex'] == 'male') & (data['Survived'] == 0)).sum()
print("dead male", male_dead)

# Count the number of male passengers who survived (Survived == 1 and Sex == 'male')
male_survived = ((data['Sex'] == 'male') & (data['Survived'] == 1)).sum()
print("alive male", male_survived)
```

```python
# Create a bar plot to show the number of male passengers who died vs survived

plt.bar(['DEAD','SURVIVED'],[male_dead,male_survived])
plt.xlabel("survival status")
plt.ylabel("NUmber of male passengeers")
plt.title("number of dead and survived malepassengers")
plt.show()
```

Pclass

```python
# Count the number of passengers in each passenger class (Pclass)
data['Pclass'].value_counts()
```

```python
# Count the number of passengers in each class (1st, 2nd, 3rd class)
# and display the total number in each class
first_class_count = (data['Pclass'] == 1).sum()
print("First class :", first_class_count)

second_class_count = (data['Pclass'] == 2).sum()
print("Second class :", second_class_count)

third_class_count = (data['Pclass'] == 3).sum()
print("Third class :", third_class_count)
```

```python
# Create a pie chart to visualize the distribution of passengers by class (1st, 2nd, 3rd)
labels = ['First class', 'Second class', 'Third class']
sizes = [first_class_count, second_class_count, third_class_count]
plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.title("Class distribution")
plt.show()
```

```python
# Plot a countplot to show survival rates (dead or survived) based on the passenger class
sns.countplot(x=data['Pclass'], hue=data['Survived'])
plt.title("Survival as per class")
plt.xlabel("Classes")
plt.ylabel("Count")
plt.show()
```

```python
# Plot kernel density estimation (KDE) plots to compare survival rates
# based on passenger class (Pclass)
plt.figure(figsize=(16,8))
sns.kdeplot(data.Pclass[data.Survived == 0], label="died",color ='red')
sns.kdeplot(data.Pclass[data.Survived == 1], label="survived", color = 'blue')
plt.title("Survival with Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```
